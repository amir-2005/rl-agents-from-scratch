```python
import gymnasium as gym
import matplotlib.pyplot as plt
from LinearSARSA import LinearSARSA
```

```python
env = gym.make("CartPole-v1")
```

```python
agent = LinearSARSA(env, learning_rate=0.003, epsilon_start=1, epsilon_decay=0.995, epsilon_min=0.01, discount_factor=0.999)
rewards = agent.train(n_episodes=3000)
```

```python
plt.figure(figsize=(10,3))
plt.bar(x=range(len(rewards)), height=rewards)
plt.title("Train Rewards")
plt.xlabel("Episode")
plt.show()
```

```python
n_test = 3
env_2 = gym.make("CartPole-v1", render_mode="human")
r = agent.test(env_2, n_test)

print(f"Rewards in each episode:")
print(r)
```
